# **Data Preprocessing**

### Handling Categorical Features

Certain categorical features (`sex`, `cp`, `fbs`, `restecg`, `exang`, `slope`, `ca`, `thal`) should be converted into dummy variables using `pd.get_dummies()` to avoid unintended ordinal interpretations.

#### Example: `sex` Feature
The `sex` column, with values `0` and `1`, is binary. Keeping it as-is might cause the algorithm to treat `0` as "lower" and `1` as "higher," which is inappropriate since gender is not ordinal. Encoding it as dummy variables ensures proper interpretation and unbiased model predictions.

```python
dataset = pd.get_dummies(df, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

## Feature Scaling

```python
dataset.columns
```

```python
standScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
# Splitting the dataset into dependent and independent features
X = dataset.drop('target', axis=1)
y = dataset['target']
```

# **Model Building**
This phase involves experimenting with three classification algorithms to assess their performance on the dataset:
1. K-Nearest Neighbors (KNN) Classifier
2. Decision Tree Classifier
3. Random Forest Classifier

## KNeighbors Classifier Model

```python
# Finding the best accuracy for knn algorithm using cross_val_score 
knn_scores = []
for i in range(1, 21):
  knn_classifier = KNeighborsClassifier(n_neighbors=i)
  cvs_scores = cross_val_score(knn_classifier, X, y, cv=10)
  knn_scores.append(round(cvs_scores.mean(),3))
```

```python
# Plotting the results of knn_scores
plt.figure(figsize=(20, 15))  # Set the figure size
plt.plot([k for k in range(1, 21)], knn_scores, color='red')  # Plot KNN scores for each K value

# Annotating each point on the plot
for i in range(1, 21):
    plt.text(i, knn_scores[i-1], f'({i}, {knn_scores[i-1]:.2f})', ha='center', va='bottom')

# Customizing the plot
plt.xticks([i for i in range(1, 21)])  # Set x-axis ticks from 1 to 20
plt.xlabel('Number of Neighbors (K)')  # Label for x-axis
plt.ylabel('Scores')  # Label for y-axis
plt.title('K Neighbors Classifier Scores for Different K Values')  # Title of the plot
plt.show()  # Display the plot
```

```python
# Training the knn classifier model with k value as 12
knn_classifier = KNeighborsClassifier(n_neighbors=12)
cvs_scores = cross_val_score(knn_classifier, X, y, cv=10)
print("KNeighbours Classifier Accuracy with K=12 is: {}%".format(round(cvs_scores.mean(), 4)*100))
```

## Decision Tree Classifier

```python
# Finding the best accuracy for decision tree algorithm using cross_val_score 
decision_scores = []
for i in range(1, 11):
  decision_classifier = DecisionTreeClassifier(max_depth=i)
  cvs_scores = cross_val_score(decision_classifier, X, y, cv=10)
  decision_scores.append(round(cvs_scores.mean(),3))
```

```python
# Plotting the results of decision_scores
plt.figure(figsize=(20, 15))  # Set the figure size
plt.plot([i for i in range(1, 11)], decision_scores, color='red')  # Plot decision tree scores for each depth value

# Annotating each point on the plot
for i in range(1, 11):
    plt.text(i, decision_scores[i-1], f'({i}, {decision_scores[i-1]:.2f})', ha='center', va='bottom')

# Customizing the plot
plt.xticks([i for i in range(1, 11)])  # Set x-axis ticks from 1 to 10 (Depth values)
plt.xlabel('Depth of Decision Tree (N)')  # Label for x-axis
plt.ylabel('Scores')  # Label for y-axis
plt.title('Decision Tree Classifier Scores for Different Depth Values')  # Title of the plot
plt.show()  # Display the plot
```

```python
# Training the decision tree classifier model with max_depth value as 3
decision_classifier = DecisionTreeClassifier(max_depth=3)
cvs_scores = cross_val_score(decision_classifier, X, y, cv=10)
print("Decision Tree Classifier Accuracy with max_depth=3 is: {}%".format(round(cvs_scores.mean(), 4)*100))
```

## Random Forest Classifier

```python
# Finding the best accuracy for random forest algorithm using cross_val_score 
forest_scores = []
for i in range(10, 101, 10):
  forest_classifier = RandomForestClassifier(n_estimators=i)
  cvs_scores = cross_val_score(forest_classifier, X, y, cv=5)
  forest_scores.append(round(cvs_scores.mean(),3))
```

```python
# Plotting the results of forest_scores
plt.figure(figsize=(20,15))
plt.plot([n for n in range(10, 101, 10)], forest_scores, color = 'red')
for i in range(1,11):
    plt.text(i*10, forest_scores[i-1], f'({i*10}, {forest_scores[i-1]:.2f})', ha='center', va='bottom')
plt.xticks([i for i in range(10, 101, 10)])
plt.xlabel('Number of Estimators (N)')
plt.ylabel('Scores')
plt.title('Random Forest Classifier Scores for Different N Values')
plt.show()  # Display the plot
```

```python
# Training the random forest classifier model with n value as 90
forest_classifier = RandomForestClassifier(n_estimators=90)
cvs_scores = cross_val_score(forest_classifier, X, y, cv=5)
print("Random Forest Classifier Accuracy with n_estimators=90 is: {}%".format(round(cvs_scores.mean(), 4)*100))
```
